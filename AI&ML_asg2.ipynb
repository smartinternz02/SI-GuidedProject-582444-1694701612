```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
df=pd.read_csv("/content/House Price India.csv")
df.head()
```


3. Perform the Below Visualizations.
● Univariate Analysis
● Bi - Variate Analysis
● Multivariate Analysis

on the above dataset

Univaraite Anlaysis

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Price', bins=20, kde=True)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Bivariate anlysis

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='number of bedrooms', y='Price')
plt.title('Scatter Plot: Price vs. Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()


```

Mulltivariate analysis

```python

numerical_columns = ['Price', 'living area', 'number of bedrooms', 'number of bathrooms']
sns.pairplot(df[numerical_columns])
plt.show()
```
